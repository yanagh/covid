```python
sortedTotalVaccinations = df2.sort_values(by=['total_vaccinations'], ascending=False)
sortedTotalVaccinations.head()
```

```python
df3 = sortedTotalVaccinations.groupby(['country'], sort=False)['vaccines','people_fully_vaccinated','total_vaccinations'].max() 
```
